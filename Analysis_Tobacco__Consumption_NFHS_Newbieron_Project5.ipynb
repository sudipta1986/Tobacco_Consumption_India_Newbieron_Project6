```python
# Import require library for creating charts.
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Import csv file for analysis.
sm=pd.read_csv('Tobacco Consumption India.csv')
sm.head()
```

```python
# For checking rows and column in data set.
sm.shape
```

```python
# Check data types.
sm.dtypes
```

```python
# We have to change the data type 'Women' and 'Men' columns but as it's a string also we have '%' in each elements so we have to separate that '%' from
# each rows then we can convert the data type. So we applied 'lambda' function fllow by 'split' function and index.
sm['Women']=sm['Women'].apply(lambda x: x.split("%")[0])
sm['Men']=sm['Men'].apply(lambda x: x.split("%")[0])
```

```python
# Now we can change the data type of 'Women' and 'Men' columns with the help of 'astype' function.
sm=sm.astype({'Women':'Float64','Men':'Float64'})
```

```python
# Check the data info.
sm.info()
```

```python
# check the data.
sm.head()
```

# **Analysis**

```python
# This is the plot size configaration.
plt.figure(figsize=(10,10))

# Now we can use plot style but it's not mandatory.
sns.set_style("whitegrid")

# In order to sort the data in ascending order.
sm_sh = sm.sort_values(by=['State'],ascending=True)

# We used seaborn to making the diagram, Y axis is define 'State' and hue or we can say legend is define 'Men' columns and mention the data set.
sn = sns.barplot(y='State',x='Men', data=sm_sh)

# We created this forloop for mention the count on each bar.
for i in sn.containers:
  sn.bar_label(i, fmt='%.1f%%', label_type='edge')

plt.xlabel("Tobacco consumption by Men in '%'") # This is title of x axis.
plt.ylabel("States") # This is title of Y axis.
plt.title("Tobacco consumption by Men of each state in India") # This is Title of Diagram.
```

After conducting an in-depth analysis of the National Family Health Survey data on tobacco consumption among men and women in each state of India, it is apparent that men are significantly more affected by smoking in comparison to women. State-wise comparisons highlight Mizoram as the top-ranking state with the highest tobacco consumption among both genders. Conversely, states such as Punjab, Chandigarh, and Himachal Pradesh exhibit the lowest rates of tobacco consumption among women in India. Notably, Chandigarh stands out as the state with the least tobacco consumption among men, followed by Punjab, Puducherry, and Kerala. The findings underscore the importance of targeted efforts to address tobacco use and its associated health risks in different states and gender-specific contexts.
